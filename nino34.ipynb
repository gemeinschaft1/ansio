```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('nino34.long.anom.data.csv')
```

```python
data
```

```python
mean=[]

for i in range(len(data)):
  mean.append(data.iloc[i,1:].mean())

plt.figure(figsize=(10,6))  # Increase the figure size

sns.lineplot(x=data.iloc[:,0].values,y=mean)
plt.title('Annual Mean Anomaly Temperature \n1870-2023')
plt.ylabel('Temperature (Celcius)')
plt.xlabel('Year')
plt.tight_layout()  # Adjust the spacing
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
monthly_mean=[]

for i in months:
  monthly_mean.append(data[i].mean())

sns.lineplot(x=months,y=monthly_mean)
plt.ylabel('Temperature (Celcius)')
#plt.gca().invert_yaxis()
plt.title('Monthly Average Anomaly Temperature 1870-2023')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 10))

data1 = data[(data.iloc[:,1:] >= 0) & (data.iloc[:,1:] < 1)].values.flatten()
data2 = data[(data.iloc[:,1:] >= 1) & (data.iloc[:,1:] < 2)].values.flatten()
data3 = data[(data.iloc[:,1:] >= 2) & (data.iloc[:,1:] < 3)].values.flatten()
data4 = data[(data.iloc[:,1:] >= -3) & (data.iloc[:,1:] < -2)].values.flatten()
data5 = data[(data.iloc[:,1:] >= -2) & (data.iloc[:,1:] < -1)].values.flatten()
data6 = data[(data.iloc[:,1:] >= -1) & (data.iloc[:,1:] < 0)].values.flatten()

data_lengths = [len(data1[~np.isnan(data1)]), len(data2[~np.isnan(data2)]), len(data3[~np.isnan(data3)]), len(data4[~np.isnan(data4)]), len(data5[~np.isnan(data5)]), len(data6[~np.isnan(data6)])]

labels = ['(0,1)', '(1,2)', '(2,3)', '(-3,-2)', '(-2,-1)', '(-1,-0)']

ax.pie(data_lengths, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'])
ax.legend(labels, loc='best')

plt.tight_layout()
plt.title('Pie Chart of Anomaly Temperature')
plt.show()
```

```python
#Frekuensi untuk setiap rentang temperatur
sns.histplot(data.iloc[:,1:].values.flatten(), binrange=(-2.5,2.57), binwidth=0.5)
plt.ylabel('Frequency')
plt.title('Frequency of Anomaly Temperature 1870-2023')
plt.xlabel('Temperature (Celcius)')
plt.show()
```

```python
import numpy as np
# Flatten your data
data_flattened = data.iloc[:,1:].values.flatten()

# Define your bins
bins = np.arange(-2.5, 4, 0.5)

# Segment and sort data values into bins
data_binned = pd.cut(data_flattened, bins)

# Count the frequency of each bin
bin_counts = data_binned.value_counts()

# Sort by bin
bin_counts = bin_counts.sort_index()

# Print bin counts
print('Frequency of Anomaly Temperature 1870-2023 for Each Bin')
print(bin_counts)
```

```python
sns.boxplot(data=data.iloc[:,1:])
plt.title('Monthly Boxplot for Anomaly (1870-2023)')
plt.ylabel('Temperature (Celcius)')
plt.xlabel('Month')
plt.show()
```

```python
# Small Multiples
year = list(range(1870, 2024))
for idx, month in enumerate(months):
  fig=plt.figure(figsize=(15,5))
  plt.plot(year, data[month])
  plt.title('Annual Time Series Anomaly on '+' '+month)
```

```python
for idx, month in enumerate(months):
  fig=plt.figure(figsize=(15,5))
  sns.histplot(data[month], kde=True, color='blue')
  ax=fig.gca()
  ax.axvline(data[month].mean(), color='#ff033e', linestyle='dashed', linewidth=2)  #Rose-Red Color indicate mean of data
  ax.axvline(data[month].median(), color='#00ffff', linestyle='dashed', linewidth=2) #Cyan indicate median of data
  plt.title('Distribusi pada Bulan'+' '+month)
```

```python
fig, ax = plt.subplots()

# Customize the colors
colors = ['#1f77b4', '#ff7f0e']

data1870_1900 = data.iloc[:31,1:].values.flatten()
data1990_2020 = data.iloc[120:-4,1:].values.flatten()
# Plot the violin plot with customized colors
violin_parts = ax.violinplot([data1870_1900, data1990_2020], showmeans=True, showmedians=True, widths=0.8, vert=False, positions=[0, 1], showextrema=False)

# Customize the violin plot elements
for pc, color in zip(violin_parts['bodies'], colors):
  pc.set_facecolor(color)
  pc.set_edgecolor('black')
  pc.set_alpha(0.7)

violin_parts['cmeans'].set_color('red')
violin_parts['cmedians'].set_color('blue')

# Add labels and titles
ax.set_yticks([0, 1])
ax.set_yticklabels(['1870-1900', '1990-2020'])
ax.set_xlabel('Temperature (Celcius)')
ax.set_title('Violin Plot for Anomaly Temperature \n(1870-1900 and 1990-2020)')

plt.show()
```
